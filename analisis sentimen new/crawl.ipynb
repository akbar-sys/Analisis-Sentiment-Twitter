```python
df = pd.read_csv("vaksin_data.csv")
#df.head()
```

```python
df["tweet"] = df["tweet"].str.lower()
print(df["tweet"].head())
```

```python
from nltk.tokenize import word_tokenize
import re
import string
```

```python
def hapus_karakter(text):
  text = str(text)
  text = text.replace('\\t', " ").replace('\\n'," ").replace('\\u', " ").replace('\\', "")  #menghapus tab, new line, back slice
  text = text.encode('ascii', 'replace').decode('ascii') #menghapus non ascii
  text = ' '.join(re.sub("([@#_#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ",text).split())  #menghapus hashtag, mention dan link
  return text.replace("http://", " ").replace("https://", " ")

df['tweet'] = df['tweet'].apply(hapus_karakter)
print(df['tweet'].head())
```

```python
def hapus_punc(text):
    text = str(text)
    return text.translate(str.maketrans('','',string.punctuation))

df['tweet'] = df['tweet'].apply(hapus_punc)
print(df['tweet'].head())
```

```python
def remove_rt(text):
  return re.sub('rt', '', text)

df['tweet'] = df['tweet'].apply(remove_rt)
print(df['tweet'].head())
```

```python
#normalisasi
norm = pd.read_excel("normalisasi_text.xlsx")
```

```python
dict_normalisasi = {}

for index, row in norm.iterrows():
    if row[0] not in dict_normalisasi:
        dict_normalisasi[row[0]]= row[1]

def normalisasi_term(data):
    return [dict_normalisasi[term] if term in dict_normalisasi else term for term in data]

dfmod['tweet_norm'] = dfmod['tweet'].apply(normalisasi_term)
dfmod['tweet_norm'].head(10)
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
def stemming(text):
  return stemmer.stem(text)

text_dict = {}

for dokumen in dfmod['tweet_norm']:
  for text in dokumen:
    if text not in text_dict:
      text_dict[text] = ' '

print(len(text_dict))
print("-------------------------")

for text in text_dict:
  text_dict[text] = stemming(text)
  print(text, " ",text_dict[text])

print(text_dict)
print("-------------------------")

def get_stemming(dokumen):
  return [text_dict[text] for text in dokumen]

dfmod['tweet_stemming'] = dfmod['tweet_norm'].apply(get_stemming)
print(dfmod['tweet_stemming'].head())
```
