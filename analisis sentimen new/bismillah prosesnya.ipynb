```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data_netral = [row['tweet'] for index,row in df.iterrows() if row["sentimen"]=='netral']

vec_netral = CountVectorizer()
d_netral = vec_netral.fit_transform(data_netral)
tdm_netral = pd.DataFrame(d_netral.toarray(), columns=vec_netral.get_feature_names())

tdm_netral
```

```python
data_negatif = [row['tweet'] for index,row in df.iterrows() if row["sentimen"]=='negatif']

vec_negatif = CountVectorizer()
d_negatif = vec_negatif.fit_transform(data_negatif)
tdm_negatif = pd.DataFrame(d_negatif.toarray(), columns=vec_negatif.get_feature_names())

tdm_negatif
```

```python
data_positif = [row['tweet'] for index,row in df.iterrows() if row["sentimen"]=='positif']

vec_positif = CountVectorizer()
d_positif = vec_positif.fit_transform(data_positif)
tdm_positif = pd.DataFrame(d_positif.toarray(), columns=vec_positif.get_feature_names())

tdm_positif
```

```python
kata_netral = vec_netral.get_feature_names();
hitung_kata_netral = d_netral.toarray().sum(axis=0)
frek_netral = dict(zip(kata_netral,hitung_kata_netral))
#frek_netral
```

```python
kata_negatif = vec_negatif.get_feature_names();
hitung_kata_negatif = d_negatif.toarray().sum(axis=0)
frek_negatif = dict(zip(kata_negatif,hitung_kata_negatif))
#frek_negatif
```

```python
kata_positif = vec_positif.get_feature_names();
hitung_kata_positif = d_positif.toarray().sum(axis=0)
frek_positif = dict(zip(kata_positif,hitung_kata_positif))
#frek_positif
```

```python
peluang_netral = []
for kata, count in zip(kata_netral,hitung_kata_netral):
    peluang_netral.append(count/len(kata_netral))
angka_net = dict(zip(kata_netral,peluang_netral))
#angka_net
```
